```python
import numpy as np

signal = np.random.rand(10, 50, 50).astype(np.float32) * 0.05

# inject a bright spherical blob
z, y, x = 5, 25, 25
signal[z-1:z+2, y-3:y+3, x-3:x+3] += 5.0

signal.shape
```

```python
cells = main(
    signal_array=signal,
    voxel_sizes=(5, 2, 2),
    soma_diameter=16
)

len(cells), cells
```
